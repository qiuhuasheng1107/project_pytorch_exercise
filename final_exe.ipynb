```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customer.csv')
df.head()
```

```python
class_name = df['Education'].value_counts().index.tolist()
class_name
```

```python
amount = df['Education'].value_counts().tolist()
amount
```

```python
colors = ['#0081a7', '#00afb9', '#fdfcdc', '#fed9b7', '#f07167']
```

```python
plt.pie(amount, labels=class_name, colors=colors, 
        autopct='%1.0f%%', radius=1.5)
plt.show()
```

Most: Graduation; Least: Basic

```python
sub_set = df[['MntFruits', 'MntMeatProducts', 'NumStorePurchases', 'NumWebPurchases']]
sub_set.head()
```

```python
coco = sub_set.corr()
```

```python
sns.heatmap(coco, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.show()
```

spendings on fruit and spendings on meat are higly correlated.
